### Random Forest Regression

#### Imports

```python
import numpy as np
import matplotlib.pyplot as py
import pandas as pd
```

#### Importing the dataset

```python
dataset = pd.read_csv("dataset/Position_Salaries.csv")
dataset
```

```python
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values
```

```python
X
```

```python
Y.values
```

#### Training the Random Forest Regression model on whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X,Y)
```

```python
regressor.predict([[6.5]])
```

#### Visualizing the Random Forest Regression Results

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid),1))
py.scatter(X, Y, color='red')
py.plot(X_grid, regressor.predict(X_grid), color='blue')
py.title("Truth, or Bluff (Random Forest Regression)")
py.xlabel("Position Level")
py.ylabel("Salary")
py.show()
```
